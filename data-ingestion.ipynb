```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]
```

```python
# Read the data from https://raw.githubusercontent.com/tantikristanti/NYC-Taxi/
prefix = 'https://github.com/tantikristanti/NYC-Taxi/releases/download/v1.0.0-yellow-alpha'
file_name = 'yellow_tripdata_2026-01.csv.gz'
url = f'{prefix}/{file_name}'
print('Read URL:', url)
try:
    df_01_2026 = pd.read_csv(url,
                            dtype=dtype,
                            parse_dates=parse_dates)
    print('Success! Shape:', df_01_2026.shape)
except Exception as e:
    print('Error:', e)
```

```python
# Display first rows
df_01_2026.head()
```

```python
# Read the CSV data in batches. The size of each batch is, for example, 100000
df_01_2026_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000 # Size of each batch; it can be adjusted based on the available memory and performance requirements.
)
```

```python
# Create Database Connection
connection_engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi_db')
```

```python
# Complete Data Ingestion Process
header = True
for df_chunk in df_01_2026_iter: 
    # First, create the structure of the table without any data
    # If the table already exists, replace it (delete the existing table and create a new one).
    if header: 
        # Create only the table schema 
        df_chunk.head(0).to_sql(
            name="yellow_taxi_data_01_2026", # the table name
            con=connection_engine,
            if_exists="replace" 
        )
        header = False 
        print("Table created")

    # If the table exists, append with data in batches.
    df_chunk.to_sql(
        name="yellow_taxi_data_01_2026",
        con=connection_engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk))
```
