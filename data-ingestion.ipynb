## Understand the Dataset

```python
# Import Libraries
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Read the data from https://github.com/tantikristanti/Datasets/
prefix = 'https://github.com/tantikristanti/Datasets/releases/download/v1.0.0-yellow-alpha'
file_name = 'yellow_tripdata_2026-01.csv.gz'
url = f'{prefix}/{file_name}'
df_01_2026 = pd.read_csv(url)
```

```python
df_01_2026.head()
```

```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]
```

```python
# Read the CSV data in batches. The size of each batch is, for example, 100000
df_01_2026_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000 # Size of each batch; it can be adjusted based on the available memory and performance requirements.
)
```
